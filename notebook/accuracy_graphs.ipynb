```python
import pandas as pd
import numpy as np
```

```python
er = pd.read_csv('oil_er.csv')
er = er.T
er = er[1:]
er.plot()
```

```python
tmp = pd.Series(np.NaN,index=np.arange(2000),dtype='float64')
er['Domain 3'] = tmp
tmp = er[3][:500].copy()
er['Domain 3'][:2000] = np.NaN
er['Domain 3'][1500:] = tmp
```

```python
er = er.rename(columns={0:'Domain 0'})
```

```python
er = er.drop([1,2,3],axis=1)
```

```python
er
```

```python
er.to_csv('oil_er.csv')
```

```python
online = pd.read_csv("oil_online.csv")
online = online.drop(['Unnamed: 0'],axis=1)
ax = online.plot(figsize=(12,8),title='Online Test')
ax.set_xlabel("Epochs")
ax.set_ylabel("Accuracy")
ax.xaxis.grid(True)
ax.yaxis.grid(True)
ax.set_ylim(30,100)
```
