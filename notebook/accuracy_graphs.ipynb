```python
import pandas as pd
```

```python
online = pd.read_csv("csv/online.csv")
online = online.drop(['Unnamed: 0'],axis=1)
ax = online.plot(figsize=(12,8),title='Online Test')
ax.set_xlabel("Epochs")
ax.set_ylabel("Accuracy")
ax.xaxis.grid(True)
ax.yaxis.grid(True)
ax.set_ylim(30,100)
```
