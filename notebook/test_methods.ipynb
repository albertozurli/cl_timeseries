```python
cnn = pd.read_csv("test_cnn.csv")
cnn = cnn.drop(['Unnamed: 0'],axis=1)
ax = cnn.plot(figsize=(12,8),title='CNN Test')
ax.set_xlabel("Epochs")
ax.set_ylabel("Accuracy")
ax.xaxis.grid(True)
ax.yaxis.grid(True)
ax.set_ylim(30,100)
```

```python
norm = pd.read_csv("norm.csv")
norm = norm.drop(['Unnamed: 0'],axis=1)
ax = norm.plot(figsize=(12,8),title='CNN Test')
ax.set_xlabel("Epochs")
ax.set_ylabel("Accuracy")
ax.xaxis.grid(True)
ax.yaxis.grid(True)
ax.set_ylim(30,100)
```

```python
drop = pd.read_csv("dropout.csv")
drop = drop.drop(['Unnamed: 0'],axis=1)
ax = drop.plot(figsize=(12,8),title='CNN Test')
ax.set_xlabel("Epochs")
ax.set_ylabel("Accuracy")
ax.xaxis.grid(True)
ax.yaxis.grid(True)
ax.set_ylim(30,100)
```

```python
df = pd.DataFrame()
index = 4
```

```python
df['Normalization'] = norm[f'Domain {index}']
df['CNN'] = cnn[f'Domain {index}']
df['Dropout'] = drop[f'Domain {index}']
```

```python
ax = df.plot(figsize=(12,8),title=f'Evaluation Domain {index}')
ax.set_xlabel("Epochs")
ax.set_ylabel("Accuracy")
ax.xaxis.grid(True)
ax.yaxis.grid(True)
ax.set_ylim(30,100)
```

```python
df['Normalization'] = norm['Mean']
df['CNN'] = cnn['Mean']
df['Dropout'] = drop['Mean']
```

```python
ax = df.plot(figsize=(12,8),title=f'Mean Evaluation')
ax.set_xlabel("Epochs")
ax.set_ylabel("Accuracy")
ax.xaxis.grid(True)
ax.yaxis.grid(True)
ax.set_ylim(30,100)
```
